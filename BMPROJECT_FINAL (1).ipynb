```python
import pandas as pd
from urllib.request import urlopen
import io

# -----------------------------
# Step 0: Define cycles and base URLs
# -----------------------------
years = ["2003", "2005", "2007", "2009", "2011", "2013", "2015", "2017"]
cycles = ["C", "D", "E", "F", "G", "H", "I", "J"]

def load_xpt(url, columns=None):
    try:
        df = pd.read_sas(url, format='xport', encoding='latin1')
        return df[columns] if columns else df
    except Exception as e:
        print(f"Failed to load {url}: {e}")
        return pd.DataFrame()

def collect_by_variable(base_template, years, cycles, columns, skip_last=False):
    urls = [
        base_template.format(y, c)
        for y, c in zip(years, cycles)
    ]
    if skip_last:
        urls = urls[:-1]  # Drop final cycle for variables like ALQ101
    dfs = [load_xpt(url, columns) for url in urls]
    return pd.concat(dfs, ignore_index=True)

# -----------------------------
# Step 1: Load datasets
# -----------------------------
demo = collect_by_variable(
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{}/DataFiles/DEMO_{}.xpt",
    years, cycles, ["SEQN", "RIDAGEYR", "RIAGENDR", "RIDRETH1"]
)

bmi = collect_by_variable(
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{}/DataFiles/BMX_{}.xpt",
    years, cycles, ["SEQN", "BMXBMI"]
)

bp = collect_by_variable(
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{}/DataFiles/BPX_{}.xpt",
    years, cycles, ["SEQN", "BPXSY1", "BPXDI1"]
)

smoking = collect_by_variable(
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{}/DataFiles/SMQ_{}.xpt",
    years, cycles, ["SEQN", "SMQ020"]
)

alcohol = collect_by_variable(
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{}/DataFiles/ALQ_{}.xpt",
    years, cycles, ["SEQN", "ALQ101"],
    skip_last=True  # Exclude 2017–2018 for ALQ101
)

# -----------------------------
# Step 2: Merge all datasets
# -----------------------------
df = demo.merge(bmi, on="SEQN", how="outer") \
         .merge(bp, on="SEQN", how="outer") \
         .merge(smoking, on="SEQN", how="outer") \
         .merge(alcohol, on="SEQN", how="outer")

# -----------------------------
# Step 3: Apply filters
# -----------------------------
filtered_df = df[
    (df["RIDRETH1"].isin([1, 2])) &        # Hispanic groups
    (df["RIDAGEYR"] >= 18) &               # Adults 18+
    (df["BMXBMI"] >= 30) &                 # Obese
    (df["BPXSY1"] >= 130) &                # High systolic BP
    (df["SMQ020"] == 1) &                  # Smoker
    (df["ALQ101"] == 1)                    # Drinker
]

# -----------------------------
# Step 4: Output shape and sample
# -----------------------------
print(f"Filtered dataset shape: {filtered_df.shape}")
print(filtered_df.head())
```

#**7. Dataset exploration**

Data Merging and Filtering Process
Data Source: NHANES 2003–2018 survey cycles (every 2 years)

Components used:

DEMO: demographics (age, race/ethnicity, BMI)

SMQ: smoking status

ALQ: alcohol consumption

BPX: blood pressure (systolic)

KIQ_U: kidney disease (self-reported, where available)

🧩Data Merging (Simple Explanation)
We combined information from different NHANES files by matching each person using their unique ID number (like a personal barcode).

We only kept people who had complete information across all the key areas we needed: age, race, BMI, smoking, drinking, and blood pressure.

For kidney disease, we added the data when it was available although some people didn’t have this part, we still kept them in the dataset.

🔍 Filtering Criteria
Selected participants who met all of the following:
Hispanic ethnicity → RIDRETH1 = 1 or 2

Age ≥ 18 years → RIDAGEYR ≥ 18

Obese → BMXBMI ≥ 30

Smoker → SMQ020 = 1 (Yes)

Drinker → ALQ101 = 1 (Yes)

High systolic blood pressure → BPXSY1 ≥ 130
 (includes pre-hypertensive and hypertensive individuals)

📊 Blood Pressure Categorization
SBP= Systolic Blood Pressure
Classified into 3 risk groups:
Normal: SBP < 120

Elevated: SBP 120–129

Hypertensive: SBP ≥ 130

🧬 Gender Comparison
After filtering, participants were grouped by gender (RIAGENDR)

For each gender group, we calculated the percentage with kidney disease

Based on KIQ022 (self-reported weak/failing kidneys)

Or lab-based markers like eGFR (ACR) or urinary creatinine

 Handle missing data
-If a variable was incomplete or missing for one or more survey cycles, we chose to exclude that variable entirely to maintain consistency and accuracy.
-This helped us ensure we were only analyzing data that was fully available and reliable across all years from 2003 to 2018.

Keywords used in imputing data: Replace

```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
import io
from scipy.stats import skew, shapiro

# NHANES URLs for BMXBMI
bmi_urls = [
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/BMX_C.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/BMX_D.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/BMX_E.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/BMX_F.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/BMX_G.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/BMX_H.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/BMX_I.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/BMX_J.xpt"
]

# Function to load BMXBMI
def load_bmi(url):
    response = urlopen(url)
    data = response.read()
    df = pd.read_sas(io.BytesIO(data), format='xport', encoding='latin1')
    if 'BMXBMI' in df.columns:
        return df[['BMXBMI']]
    else:
        return None

# Load and combine all data
frames = [load_bmi(url) for url in bmi_urls]
bmi_df = pd.concat([df for df in frames if df is not None], ignore_index=True)

# Drop missing values
bmi_df = bmi_df.dropna(subset=['BMXBMI'])

# Extract BMI data
bmi_data = bmi_df['BMXBMI']

# Descriptive statistics
summary_stats = {
    'Count': bmi_data.count(),
    'Mean': bmi_data.mean(),
    'Median': bmi_data.median(),
    'Standard Deviation': bmi_data.std(),
    '25th Percentile': bmi_data.quantile(0.25),
    '75th Percentile': bmi_data.quantile(0.75),
    'Minimum': bmi_data.min(),
    'Maximum': bmi_data.max(),
    'Skewness': skew(bmi_data)
}

# Shapiro-Wilk normality test (for n <= 5000)
sample_for_test = bmi_data.sample(n=5000, random_state=42) if len(bmi_data) > 5000 else bmi_data
shapiro_stat, shapiro_p = shapiro(sample_for_test)

# Output
print("Body Mass Index (BMXBMI) Descriptive Statistics:")
for stat, value in summary_stats.items():
    print(f"{stat}: {value:.2f}")

print(f"\nShapiro-Wilk Test Statistic: {shapiro_stat:.4f}")
print(f"Shapiro-Wilk p-value: {shapiro_p:.4g}")
print("Normality:", "Assumed" if shapiro_p > 0.05 else "Not normally distributed")
```

```python
import pandas as pd
from urllib.request import urlopen
import io
from scipy.stats import shapiro

# NHANES ACR file info (label, URL)
acr_files = [
    ("2003–2004", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/L16_C.xpt"),
    ("2005–2006", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/ALB_CR_D.xpt"),
    ("2007–2008", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/ALB_CR_E.xpt"),
    ("2009–2010", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/ALB_CR_F.xpt"),
    ("2011–2012", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/ALB_CR_G.xpt"),
    ("2013–2014", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/ALB_CR_H.xpt"),
    ("2015–2016", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/ALB_CR_I.xpt"),
    ("2017–2018", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/ALB_CR_J.xpt")
]

# Load and compute ACR = (URXUMA / URXUCR) * 100
acr_list = []
for label, url in acr_files:
    try:
        resp = urlopen(url)
        df = pd.read_sas(io.BytesIO(resp.read()), format='xport', encoding='latin1')
        if "URXUMA" in df.columns and "URXUCR" in df.columns:
            valid = df[["URXUMA", "URXUCR"]].dropna()
            acr = (valid["URXUMA"] / valid["URXUCR"]) * 100
            acr_list.append(acr)
    except Exception as e:
        print(f"Error loading {label}: {e}")

# Combine all ACR values
acr_values = pd.concat(acr_list, ignore_index=True)

# Shapiro-Wilk test (sample if n > 5000)
sample = acr_values.sample(5000, random_state=42) if len(acr_values) > 5000 else acr_values
stat, p = shapiro(sample)

# Print results
print(f"Shapiro-Wilk Test for ACR (n={len(sample)}):")
print(f"Statistic: {stat:.5f}")
print(f"P-value: {p:.5e}")

if p < 0.05:
    print("\nConclusion: ACR is not normally distributed (p < 0.05).")
else:
    print("\nConclusion: ACR appears normally distributed (p ≥ 0.05).")
```

Q-Q plot for ACR

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from urllib.request import urlopen
import io

# -----------------------------
# Step 1: Define ACR URLs
# -----------------------------
acr_urls = [
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/L16_C.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/ALB_CR_D.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/ALB_CR_E.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/ALB_CR_F.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/ALB_CR_G.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/ALB_CR_H.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/ALB_CR_I.xpt",
    "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/ALB_CR_J.xpt"
]

# -----------------------------
# Step 2: Load and compute ACR
# -----------------------------
acr_frames = []
for url in acr_urls:
    with urlopen(url) as response:
        df = pd.read_sas(io.BytesIO(response.read()), format="xport", encoding="latin1")
        if {"URXUMA", "URXUCR", "SEQN"}.issubset(df.columns):
            df = df[["SEQN", "URXUMA", "URXUCR"]].dropna()
            df = df[df["URXUCR"] > 0]
            df["ACR"] = (df["URXUMA"] / df["URXUCR"]) * 100
            acr_frames.append(df[["SEQN", "ACR"]])

# Combine all years
acr_data = pd.concat(acr_frames, ignore_index=True)

# Drop invalid or zero/negative ACR values (log-based plots would crash otherwise)
acr_values = acr_data["ACR"].dropna()
acr_values = acr_values[acr_values > 0]

# -----------------------------
# Step 3: Create Q–Q Plot
# -----------------------------
plt.figure(figsize=(8, 6))
stats.probplot(acr_values, dist="norm", plot=plt)
plt.ylim(-500, 3000)  # Set y-axis from 0 to 500
plt.title("Q–Q Plot of Albumin-to-Creatinine Ratio (ACR)\nNHANES 2003–2018")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles (ACR)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

* Strong deviation from the red line at the upper end (extreme right tail).

* Indicates severe positive skewness — a large number of small ACR values, with a few extremely high ones.

* The distribution is far from normal, even more so than URXUCR.

* Reflects that most participants have low ACR, but a small proportion may have very high levels, indicating possible kidney damage or albuminuria.

* Strongly suggests the need for data transformation (e.g. log(ACR)) or robust statistical methods.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pyreadstat
import requests
from io import BytesIO
import tempfile
import os

sns.set(style="whitegrid")

# Define base URL & years
base = "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public"
cycles = ["2003", "2005", "2007", "2009", "2011", "2013", "2015", "2017"]
suffixes = ["C", "D", "E", "F", "G", "H", "I", "J"]

# Define the file names for each dataset type, using the cycle suffix
file_names = {
    'bmi': 'BMX_{}.xpt',
    'bp': 'BPX_{}.xpt',
    'demo': 'DEMO_{}.xpt',
    'smoking': 'SMQ_{}.xpt',
    'drinking': 'ALQ_{}.xpt',
    'lab_cr': 'ALB_CR_{}.xpt', # This is the general pattern, specific exceptions handled below
    'lab_as': 'UAS_{}.xpt',   # This is the general pattern, specific exceptions handled below
}

# Handle specific file name variations for ALB_CR and UAS
special_files = {
    'lab_cr': {
        'C': 'L16_C.xpt',
    },
    'lab_as': {
        'C': 'L06UAS_C.xpt',
        'H': 'UTAS_H.xpt',
        'I': 'UTAS_I.xpt',
        'J': 'UTAS_J.xpt',
    }
}


# Load XPT files
def load_concat(dataset_key, years_full, years_suffix, rename_map):
    dfs = []
    for i, suff in enumerate(years_suffix):
        year = years_full[i]
        # Determine the correct file name for the current cycle
        if dataset_key in special_files and suff in special_files[dataset_key]:
            file_name = special_files[dataset_key][suff]
        elif dataset_key in file_names:
             file_name = file_names[dataset_key].format(suff)
        else:
            print(f"Unknown dataset key: {dataset_key}. Skipping.")
            continue


        url = f"{base}/{year}/DataFiles/{file_name}"

        try:
            response = requests.get(url)
            response.raise_for_status() # Raise an exception for bad status codes

            # Use a temporary file to read the xpt data
            with tempfile.NamedTemporaryFile(delete=False) as tmp_file:
                tmp_file.write(response.content)
                tmp_file_path = tmp_file.name

            try:
                df, _ = pyreadstat.read_xport(tmp_file_path)
                cols_to_select = list(rename_map.keys())
                existing_cols = [c for c in cols_to_select if c in df.columns]

                # Handle the specific drinking variable change in 2017-2018
                if dataset_key == 'drinking' and suff == 'J' and 'ALQ120Q' in df.columns:
                     # If ALQ120Q exists and ALQ101 does not (or we prioritize ALQ120Q for this cycle)
                     if 'ALQ101' not in existing_cols:
                          rename_map_current = rename_map.copy()
                          # Replace ALQ101 mapping with ALQ120Q for this cycle
                          if 'ALQ101' in rename_map_current:
                              rename_map_current['ALQ120Q'] = rename_map_current.pop('ALQ101')
                              cols_to_select = list(rename_map_current.keys())
                              existing_cols = [c for c in cols_to_select if c in df.columns]
                              # Note: We don't permanently change rename_map here, just for this cycle's loading


                if not existing_cols or (len(existing_cols) != len(cols_to_select) and 'SEQN' in cols_to_select):
                    print(f"Skipping {url}: Required columns {cols_to_select} not all found. Found: {existing_cols}")
                    continue

                # Select and rename the columns
                df_subset = df[existing_cols].rename(columns=rename_map)

                # Handle Drinking mapping consistently during loading
                if dataset_key == 'drinking' and 'Drinking' in df_subset.columns:
                     if suff == 'J':
                         # For 2017-2018, ALQ120Q is number of days (now mapped to 'Drinking')
                         # Map >0 to 'Yes', 0 to 'No', others (7, 9 from original ALQ120Q if they exist) to appropriate strings
                         df_subset['Drinking'] = df_subset['Drinking'].apply(lambda x: 'Yes' if x > 0 else ('No' if x == 0 else ( {7:'Refused', 9:"Don't Know"}.get(x, np.nan) )) )
                     else:
                          # For pre-2017, ALQ101 is Yes/No/Refused/Don't Know (now mapped to 'Drinking')
                          df_subset['Drinking'] = df_subset['Drinking'].map({1.0:'Yes', 2.0:'No', 7.0:'Refused', 9.0:"Don't Know"})


                dfs.append(df_subset)


            except pyreadstat.PyreadstatError as e:
                print(f"Error reading {url} with pyreadstat: {e}. Skipping this file.")
            except Exception as e:
                print(f"Error processing file {url}: {e}. Skipping this file.")
            finally:
                # Clean up the temporary file
                os.remove(tmp_file_path)

        except requests.exceptions.RequestException as e:
            print(f"Failed to download {url}: {e}. Skipping this file.")
        except Exception as e:
            print(f"Error processing URL or response for {url}: {e}. Skipping this file.")


    return pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()


# Define rename maps
rename_maps = {
    'bmi': {'BMXBMI': 'BMI',      'SEQN': 'SEQN'},
    'bp': {'BPXSY1': 'SysBP',    'SEQN': 'SEQN'},
    'demo': {'RIDAGEYR': 'Age', 'RIAGENDR': 'Gender', 'RIDRETH1': 'RaceEth', 'SEQN':'SEQN'},
    'smoking': {'SMQ020': 'Smoking',  'SEQN':'SEQN'},
    # Initial map for drinking, adjustment and mapping handled in load_concat
    'drinking': {'ALQ101':'Drinking','SEQN':'SEQN'},
    'lab_cr': {'URXUCR':'Creatinine','URXUMA':'Albumin','SEQN':'SEQN'},
    'lab_as': {'URXUAS':'TotAs','SEQN':'SEQN'}
}


# Load datasets
df_bmi = load_concat('bmi', cycles, suffixes, rename_maps['bmi'])
df_bp = load_concat('bp', cycles, suffixes, rename_maps['bp'])
df_demo = load_concat('demo', cycles, suffixes, rename_maps['demo'])
df_smoking = load_concat('smoking', cycles, suffixes, rename_maps['smoking'])
df_drinking = load_concat('drinking', cycles, suffixes, rename_maps['drinking']) # load_concat now handles mapping
df_cr = load_concat('lab_cr', cycles, suffixes, rename_maps['lab_cr'])
df_as = load_concat('lab_as', cycles, suffixes, rename_maps['lab_as'])


# Merge
from functools import reduce

dataframes_to_merge = [df_demo, df_bmi, df_bp, df_smoking, df_drinking, df_cr, df_as]
# Filter out empty dataframes before merging
dataframes_to_merge = [df for df in dataframes_to_merge if not df.empty]


if dataframes_to_merge:
    # Start with the first non-empty dataframe
    df = dataframes_to_merge[0]
    # Merge the rest
    for next_df in dataframes_to_merge[1:]:
        df = pd.merge(df, next_df, on="SEQN", how="outer") # Use outer merge to keep all SEQN from loaded files
else:
    df = pd.DataFrame() # Create an empty dataframe if no data was loaded


# Compute ACR
if 'Albumin' in df.columns and 'Creatinine' in df.columns:
    # Avoid division by zero by replacing 0 creatinine with NaN before division
    df['Creatinine_safe'] = df['Creatinine'].replace(0, np.nan)
    df['ACR'] = df['Albumin'] / df['Creatinine_safe']
    df = df.drop(columns=['Creatinine_safe']) # Drop the temporary column
else:
    df['ACR'] = np.nan # Add ACR column with NaN if not calculable


# Remap categories
# Ensure columns exist before mapping and handle potential NaN values in the original column
if 'Smoking' in df.columns:
    # Map SMQ020: 1=Yes, 2=No, 7=Refused, 9=Don't Know. Keep 7 and 9 as they are valid responses.
    smoking_map_full = {1.0:'Yes', 2.0:'No', 7.0:'Refused', 9.0:"Don't Know"}
    # Use .map() and fillna('') or pd.NA to handle missing values if needed, here we let NaN remain NaN
    df['Smoking'] = df['Smoking'].map(smoking_map_full)

# Drinking mapping is now handled in load_concat

if 'Gender' in df.columns:
    # Map RIAGENDR: 1=Male, 2=Female
    gender_map_full = {1.0:'Male', 2.0:'Female'}
    df['Gender'] = df['Gender'].map(gender_map_full)

if 'RaceEth' in df.columns:
    # Map RIDRETH1: 1-5
    race_map_full = {
        1.0:'Mexican American', 2.0:'Other Hispanic', 3.0:'Non-Hisp White',
        4.0:'Non-Hisp Black', 5.0:'Other Race'
    }
    df['RaceEth'] = df['RaceEth'].map(race_map_full)

# Plot grouped bar charts
def plot_grouped_bar(dep_var, cat_x, cat_hue):
    # Ensure columns exist before plotting
    if not all(col in df.columns for col in [dep_var, cat_x, cat_hue]):
        print(f"Skipping plot for {dep_var} by {cat_x} and {cat_hue}: One or more columns not found.")
        return

    plt.figure(figsize=(10, 6))
    # Drop rows with NaN values in the required columns for plotting
    temp_df = df.dropna(subset=[dep_var, cat_x, cat_hue])

    # Check if temp_df is empty after dropping NaNs
    if temp_df.empty:
        print(f"Skipping plot for {dep_var} by {cat_x} and {cat_hue}: No valid data points after dropping NaNs.")
        return

    sns.barplot(data=temp_df, x=cat_x, y=dep_var, hue=cat_hue, errorbar='ci', palette='viridis')
    plt.title(f"{dep_var} by {cat_x} and {cat_hue}")
    plt.ylabel(f"Mean {dep_var}")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Plot all dependent variables grouped by two categorical IVs
dependent_vars = ['Creatinine', 'ACR', 'TotAs']
groupings = [('RaceEth', 'Gender'), ('Smoking', 'Drinking')]

# Check if the combined dataframe is not empty before attempting to plot
if not df.empty:
    for dv in dependent_vars:
        for x, hue in groupings:
            plot_grouped_bar(dv, x, hue)
else:
    print("No data loaded into df. Cannot generate plots.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# NHANES cycles and file URLs
files = {
    '2003-2004': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/L16_C.xpt',
    '2005-2006': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/ALB_CR_D.xpt',
    '2007-2008': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/ALB_CR_E.xpt',
    '2009-2010': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/ALB_CR_F.xpt',
    '2011-2012': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/ALB_CR_G.xpt',
    '2013-2014': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/ALB_CR_H.xpt',
    '2015-2016': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/ALB_CR_I.xpt',
    '2017-2018': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/ALB_CR_J.xpt'
}

# Store results
median_creatinine = []

for cycle, url in files.items():
    print(f"Loading {cycle}...")
    df = pd.read_sas(url)

    if 'URXUCR' in df.columns:
        median_value = df['URXUCR'].median(skipna=True)
        median_creatinine.append({'Cycle': cycle, 'Median_Creatinine': median_value})
    else:
        print(f"Warning: URXUCR not found in {cycle}")
        median_creatinine.append({'Cycle': cycle, 'Median_Creatinine': None})

# Convert to DataFrame
median_df = pd.DataFrame(median_creatinine)

# Plot
plt.figure(figsize=(10,6))
plt.plot(median_df['Cycle'], median_df['Median_Creatinine'], marker='o', linestyle='-', color='green')
plt.xlabel('NHANES Cycle')
plt.ylabel('Median Urinary Creatinine (mg/dL)')
plt.title('Median Urinary Creatinine Across NHANES Cycles (2003-2018)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

- The overall trend from 2003 to 2018 is a decline from around 124.0mg/dL to 109.0mg/dL
- However, there was an increase from 2009 to 2012 from around 107.0mg/dL to 110.0mg/dL
- There is also an increase from 2015 to 2018 from around 102.5mg/dL to 109.0mg/dL

Annual Trend for median ACR

```python
import pandas as pd
import matplotlib.pyplot as plt

# NHANES file URLs
files = {
    '2003-2004': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/L16_C.xpt',
    '2005-2006': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/ALB_CR_D.xpt',
    '2007-2008': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/ALB_CR_E.xpt',
    '2009-2010': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/ALB_CR_F.xpt',
    '2011-2012': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/ALB_CR_G.xpt',
    '2013-2014': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/ALB_CR_H.xpt',
    '2015-2016': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/ALB_CR_I.xpt',
    '2017-2018': 'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/ALB_CR_J.xpt'
}

# Store median ACR values
median_acr = []

for cycle, url in files.items():
    print(f"Loading {cycle}...")
    df = pd.read_sas(url)

    if 'URXUMA' in df.columns and 'URXUCR' in df.columns:
        # Filter and make a copy to avoid SettingWithCopyWarning
        df_valid = df[
            (df['URXUCR'] > 0) &
            (~df['URXUMA'].isna()) &
            (~df['URXUCR'].isna())
        ].copy()

        # Compute ACR (mg/g)
        df_valid['ACR'] = df_valid['URXUMA'] / df_valid['URXUCR']

        # Compute median ACR
        median_val = df_valid['ACR'].median()
        median_acr.append({'Cycle': cycle, 'Median_ACR': median_val})
    else:
        print(f"Warning: URXUMA or URXUCR not found in {cycle}")
        median_acr.append({'Cycle': cycle, 'Median_ACR': None})

# Convert to DataFrame
median_df = pd.DataFrame(median_acr)

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(median_df['Cycle'], median_df['Median_ACR'], marker='o', linestyle='-', color='darkorange')
plt.xlabel('NHANES Cycle')
plt.ylabel('Median Albumin-to-Creatinine Ratio (mg/g)')
plt.title('Median ACR by NHANES Cycle (2003-2018)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

- The overall trend from 2003 to 2018 is an increase from 0.0655mg/g to 0.0830mg/g
- However there was a drop from 0.0775mg/g in 2007 to around 0.0700mg/g in 2009
- There was then an increase from 0.0700mg/g in 2009 to 0.0755mg/g in 2011


Annual Trend for mean BMI

```python
import pandas as pd
import matplotlib.pyplot as plt
from urllib.request import urlopen
import io

# NHANES BMXBMI file info (Cycle Label, URL)
bmi_info = [
    ("2003–2004", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/BMX_C.xpt"),
    ("2005–2006", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/BMX_D.xpt"),
    ("2007–2008", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/BMX_E.xpt"),
    ("2009–2010", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/BMX_F.xpt"),
    ("2011–2012", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/BMX_G.xpt"),
    ("2013–2014", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/BMX_H.xpt"),
    ("2015–2016", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/BMX_I.xpt"),
    ("2017–2018", "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/BMX_J.xpt"),
]

# Collect medians per cycle
years = []
medians = []

for label, url in bmi_info:
    try:
        resp = urlopen(url)
        df = pd.read_sas(io.BytesIO(resp.read()), format="xport", encoding="latin1")
        if "BMXBMI" in df.columns:
            median_val = df["BMXBMI"].dropna().median()
            medians.append(median_val)
            years.append(label)
        else:
            print(f"BMXBMI missing in {label}")
    except Exception as e:
        print(f"Error reading {label}: {e}")

# Plot
plt.figure(figsize=(10, 6))
plt.plot(years, medians, marker="o", linestyle="-", color="royalblue")
plt.title("Trend in Median Body Mass Index (BMXBMI)\nNHANES 2003–2018")
plt.xlabel("NHANES Cycle")
plt.ylabel("Median BMI (kg/m²)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

The overall trend from 2003 to 2018 is slight increase from 24.3kg/m^2 to 25.8kg/m^2

There was a constant median BMI from 2003-2004 to 2005-2006 at 24.3kg/m^2

There was a sharp drop in median BMI from 25.35kg/m^2 to 24.5kg/m^2 in 2009-2010 to 2011-2012


Annual Trend for median Systolic blood pressure

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
import io
import pyreadstat
import tempfile
import os

# Mapping of survey years to their respective URLs
nhanes_cycles = {
    "2003-2004": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/BPX_C.xpt",
    "2005-2006": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/BPX_D.xpt",
    "2007-2008": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/BPX_E.xpt",
    "2009-2010": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/BPX_F.xpt",
    "2011-2012": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/BPX_G.xpt",
    "2013-2014": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/BPX_H.xpt",
    "2015-2016": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/BPX_I.xpt",
    "2017-2018": "https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/BPX_J.xpt",
}

# Store results
results = []

# Loop through each cycle and process the data
for cycle, url in nhanes_cycles.items():
    print(f"Processing: {cycle}")

    # Download the file
    with urlopen(url) as response:
        data_bytes = response.read()

    # Use a temporary file to write the downloaded content
    with tempfile.NamedTemporaryFile(delete=False) as tmp_file:
        tmp_file.write(data_bytes)
        tmp_file_path = tmp_file.name

    try:
        # Read from the temporary file using pyreadstat
        df, meta = pyreadstat.read_xport(tmp_file_path)

        # Keep only non-missing systolic values
        df = df[['BPXSY1']].dropna()

        # Calculate median
        median_sbp = df['BPXSY1'].median()

        # Append results
        results.append({"Cycle": cycle, "Median_SBP": median_sbp})

    except Exception as e:
        print(f"Error processing {cycle}: {e}")
    finally:
        # Clean up the temporary file
        os.remove(tmp_file_path)


# Convert to DataFrame
trend_df = pd.DataFrame(results)

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=trend_df, x="Cycle", y="Median_SBP", marker="o", linewidth=2)
plt.title("Median Systolic Blood Pressure (BPXSY1) by NHANES Cycle (2003–2018)")
plt.xlabel("NHANES Cycle")
plt.ylabel("Median Systolic BP (mm Hg)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

The overall trend from 2003 to 2018 is an increase from 116.0mm Hg to 118.00mm Hg

The median systolic BP remained at 116.00mm Hg in years 2003-2004 to 2005-2006 and 2009-2010 to 2013-2014.

The median systolic BP also remained at 118.00mm Hg in 2015-2016 to 2017-2018


Percentage of sampled persons diagnosed with your disease (based on the answers in questionnaire section)


```python
import pandas as pd

# -----------------------------
# Step 0: Define NHANES cycles and years
# -----------------------------
years_all = ["2003", "2005", "2007", "2009", "2011", "2013", "2015", "2017"]
cycles_all = ["C", "D", "E", "F", "G", "H", "I", "J"]

# ALQ101 (Drinking) — Exclude 2017–2018
alq_years = ["2003", "2005", "2007", "2009", "2011", "2013", "2015"]
alq_cycles = ["C", "D", "E", "F", "G", "H", "I"]

# -----------------------------
# Step 1: Import Demographics (2003–2018)
# -----------------------------
demo_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/DEMO_{c}.xpt"
             for y, c in zip(years_all, cycles_all)]
df_demo = pd.concat([pd.read_sas(url, format='xport') for url in demo_urls], ignore_index=True)
df_demo = df_demo[["SEQN", "RIDAGEYR", "RIAGENDR", "RIDRETH1"]]  # Age, Gender, Race

# -----------------------------
# Step 2: Import ALQ101 (2003–2016 only)
# -----------------------------
alq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/ALQ_{c}.xpt"
            for y, c in zip(alq_years, alq_cycles)]
df_alq = pd.concat([pd.read_sas(url, format='xport') for url in alq_urls], ignore_index=True)
df_alq = df_alq[["SEQN", "ALQ101"]]

# -----------------------------
# Step 3: Import BMI (BMXBMI)
# -----------------------------
bmx_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BMX_{c}.xpt"
            for y, c in zip(years_all, cycles_all)]
df_bmx = pd.concat([pd.read_sas(url, format='xport') for url in bmx_urls], ignore_index=True)
df_bmx = df_bmx[["SEQN", "BMXBMI"]]

# -----------------------------
# Step 4: Import Blood Pressure (BPXSY1)
# -----------------------------
bpx_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/BPX_{c}.xpt"
            for y, c in zip(years_all, cycles_all)]
df_bpx = pd.concat([pd.read_sas(url, format='xport') for url in bpx_urls], ignore_index=True)
df_bpx = df_bpx[["SEQN", "BPXSY1"]]

# -----------------------------
# Step 5: Import Smoking Status (SMQ020)
# -----------------------------
smq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/SMQ_{c}.xpt"
            for y, c in zip(years_all, cycles_all)]
df_smq = pd.concat([pd.read_sas(url, format='xport') for url in smq_urls], ignore_index=True)
df_smq = df_smq[["SEQN", "SMQ020"]]

# -----------------------------
# Step 6: Merge all datasets
# -----------------------------
df = df_demo.merge(df_alq, on="SEQN", how="inner")
df = df.merge(df_bmx, on="SEQN", how="inner")
df = df.merge(df_bpx, on="SEQN", how="inner")
df = df.merge(df_smq, on="SEQN", how="inner")

# -----------------------------
# Step 7: Apply final filters
# -----------------------------
df_filtered = df[
    (df["RIDRETH1"].isin([1, 2])) &        # Hispanic (Mexican American or Other Hispanic)
    (df["RIDAGEYR"] >= 18) &               # Age 18+
    (df["BMXBMI"] >= 30) &                 # Obese
    (df["BPXSY1"] >= 130) &                # High systolic BP
    (df["SMQ020"] == 1) &                  # Current smoker
    (df["ALQ101"] == 1)                    # Current drinker
]

# -----------------------------
# Output
# -----------------------------
print(df_filtered.head())
print(f"Final filtered sample size: {df_filtered.shape[0]}")
```

```python
# -----------------------------
# Step 8: Import Kidney Diagnosis (MCQ160K from MCQ file)
# -----------------------------
mcq_urls = [f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{y}/DataFiles/MCQ_{c}.xpt"
            for y, c in zip(years_all, cycles_all)]
df_mcq = pd.concat([pd.read_sas(url, format='xport') for url in mcq_urls], ignore_index=True)
df_mcq = df_mcq[["SEQN", "MCQ160K"]]  # 1 = Yes, 2 = No, 7/9 = Refused/Don't Know

# -----------------------------
# Step 9: Merge diagnosis info
# -----------------------------
df_filtered = df_filtered.merge(df_mcq, on="SEQN", how="inner")
```

```python
# -----------------------------
# Step 10: Calculate % diagnosed with kidney disease
# -----------------------------
total_n = df_filtered.shape[0]
diagnosed_n = df_filtered[df_filtered["MCQ160K"] == 1].shape[0]  # Answered "Yes"
diagnosed_pct = (diagnosed_n / total_n) * 100

print(f"Total filtered sample size: {total_n}")
print(f"Number diagnosed with kidney disease: {diagnosed_n}")
print(f"Percentage diagnosed: {diagnosed_pct:.2f}%")
```

SMART Objective 1 (Mean Creatinine Level)

```python
import pandas as pd
import requests
from io import BytesIO
import os

# --------------------
# Define reusable loader
# --------------------
def load_xpt(url):
    r = requests.get(url)
    return pd.read_sas(BytesIO(r.content), format='xport')

# --------------------
# Step 1: Load or create filtered demographics (Hispanic adults 18+)
# --------------------
demo_cache = "filtered_nhanes_2003_2018.csv"

if os.path.exists(demo_cache):
    print("📂 Loading cached demographic data...")
    filtered = pd.read_csv(demo_cache)
else:
    print("🧪 Creating filtered demographics...")

    demo_urls = [
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/DEMO_C.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/DEMO_D.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/DEMO_E.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/DEMO_F.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/DEMO_G.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/DEMO_H.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/DEMO_I.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/DEMO_J.xpt'
    ]

    demo_data = pd.concat([load_xpt(url) for url in demo_urls], ignore_index=True)

    filtered = demo_data[
        (demo_data['RIDAGEYR'] >= 18) &
        (demo_data['RIDRETH1'].isin([1, 2]))  # Hispanic
    ].copy()

    filtered['Gender'] = filtered['RIAGENDR'].map({1: 'Male', 2: 'Female'})
    filtered['Ethnicity'] = filtered['RIDRETH1'].map({1: 'Mexican American', 2: 'Other Hispanic'})

    filtered.to_csv(demo_cache, index=False)
    print(f"✅ Filtered and saved to {demo_cache}")

# --------------------
# Step 2: Load and merge urinary creatinine data (URXUCR)
# --------------------
creatinine_urls = [
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/L16_C.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/ALB_CR_D.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/ALB_CR_E.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/ALB_CR_F.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/ALB_CR_G.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/ALB_CR_H.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/ALB_CR_I.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/ALB_CR_J.xpt'
]

creatinine_data = pd.concat(
    [load_xpt(url)[['SEQN', 'URXUCR']] for url in creatinine_urls],
    ignore_index=True
)

# Merge with filtered demographics
filtered_with_creatinine = pd.merge(filtered, creatinine_data, on='SEQN', how='left')

# --------------------
# Step 3: Compute and display mean creatinine by gender
# --------------------
mean_creatinine = (
    filtered_with_creatinine
    .groupby('Gender')['URXUCR']
    .mean()
    .round(2)
    .dropna()
)

print("\n📊 Mean Urinary Creatinine Levels by Gender (mg/dL):")
print(mean_creatinine)
```

Males under our sample have slightly higher mean creatinine levels at 140.45mg/dL. This could be due to the presence of the CKD cases within the 371 sample population hence pulling up the mean. This shows there could be slight relations between mean creatinine levels and CKD prevalence

SMART Objective 2 (Smoking Status)

```python
import pandas as pd
import numpy as np

# Step 1: Load creatinine + albumin data (URXUCR, URXUMA)
alb_cr_urls = [
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/L16_C.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/ALB_CR_D.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/ALB_CR_E.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/ALB_CR_F.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/ALB_CR_G.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/ALB_CR_H.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/ALB_CR_I.xpt',
    'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/ALB_CR_J.xpt'
]

alb_cr_data = pd.concat([
    load_xpt(url)[['SEQN', 'URXUCR', 'URXUMA']] for url in alb_cr_urls
], ignore_index=True)

# Step 2: Merge with filtered demographic + smoking data
smoking_data = pd.concat([
    load_xpt(url)[['SEQN', 'SMQ020']] for url in [
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2003/DataFiles/SMQ_C.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2005/DataFiles/SMQ_D.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2007/DataFiles/SMQ_E.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2009/DataFiles/SMQ_F.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2011/DataFiles/SMQ_G.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2013/DataFiles/SMQ_H.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2015/DataFiles/SMQ_I.xpt',
        'https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2017/DataFiles/SMQ_J.xpt'
    ]
], ignore_index=True)

# Merge all data
merged = filtered.merge(alb_cr_data, on='SEQN', how='left')
merged = merged.merge(smoking_data, on='SEQN', how='left')

# Step 3: Compute ACR (mg/g)
merged['ACR'] = merged['URXUMA'] / merged['URXUCR']

# Step 4: Define CKD (ACR ≥ 30 mg/g)
merged['CKD'] = np.where(merged['ACR'] >= 30, 1, 0)

# Step 5: Map smoking and gender
merged['SmokingStatus'] = merged['SMQ020'].map({1: 'Smoker', 2: 'Non-Smoker'})
merged = merged[merged['SmokingStatus'].isin(['Smoker', 'Non-Smoker'])]

# Step 6: Count CKD cases by gender and smoking
ckd_counts = merged.groupby(['Gender', 'SmokingStatus'])['CKD'].agg(['sum', 'count'])
ckd_counts['Percentage'] = (ckd_counts['sum'] / ckd_counts['count'] * 100).round(1)
ckd_counts.columns = ['CKD Cases', 'Total', 'CKD %']

print("🩺 CKD Cases by Gender and Smoking Status:")
print(ckd_counts)
```

For smoking we can see that among the whole population of U.S Hispanic adults ages 18+, for both genders, smoking increases the CKD percentages. For females its 0.3 to 0.4 and for males its 0.2 to 0.3. hence concluding a slight but present relation between smoking and CKD prevalence
